# <a id='part_a'>PART A</a>
### 2 scatter plots for variables a & b, where  a is normally distributed and b is uniformly distributed

> ## Scatter Plot - Normally Distributed

```python
color = [ "red", "blue", "green", "grey", "purple", "orange", "black","violet" ] # list of colors used to generate a random list of 1000 colors names
rand_colors = [random.choice(color) for i in range(1000)] # randomly choose a color a thousand times into an array

# generate x axis values from one to one thousand
x = list(range(0,1000))

# a normal distribution has a mean of 0 and a standard deviation of 1
mu = 0 # mean
sigma = 1 #standard deviation

# generate normal distribution
a = np.random.normal(mu, sigma, 1000)

# generate uniform distribution
b = np.random.uniform(0,1,1000)

#plot values using random colors array and normal distribution values
plt.scatter(x, a,color=rand_colors);
plt.title('Normal Distribution'); #add title
plt.ylabel('values'); #add y axis label
```

> ## Scatter Plot - Uniformly Distributed

```python
#plot values using random colors array and uniform distribution values
plt.scatter(x, b,color=rand_colors)
plt.title('Uniform Distribution');
plt.ylabel('values');
```

# <a id='part_a'>PART B</a>
### 2 histograms for normally distributed variable x a) showing density   b) without density, just count    - also vary bins

> ### A) Density Plots

```python
## subplot(2,2,1)
## this creates 4 subplots (2 rows and 2 columns)
## 1 at the end means we are in the first subplot.. then...
plt.suptitle('Normal Distribution Density Plot')

plt.subplot(2,2,1)
plt.hist(a, density=True, color = 'blue', edgecolor = 'black',bins = 10);
plt.title('10 Bins')

plt.subplot(2,2,2)
plt.hist(a, density=True, color = 'blue', edgecolor = 'black',bins = 100);
plt.title('100 Bins')

plt.subplot(2,2,3)
plt.hist(a, density=True, color = 'blue', edgecolor = 'black',bins = 200);
plt.title('200 Bins')

plt.subplot(2,2,4)
plt.hist(a, density=True, color = 'blue', edgecolor = 'black',bins = 500);
plt.title('500 Bins')

# need tight_layout to keep plots and title spaced equally
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) #needed to prevent title overlapping
plt.show();
```

> ### B) Frequency Plots

```python
#array to hold bin sizes
bins = [10,100,200,500]

#loop 4 times based on bins array
for i in range(0, len(bins)):
    plt.subplot(2,2,i+1)
    plt.hist(a, density=False, color = 'blue', edgecolor = 'black',bins = bins[i]);
    plt.title(str(bins[i]) + ' Bins')

plt.suptitle('Normal Distribution Frequency Plot')    
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) #needed to prevent title overlapping    
plt.show();
```
